```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
%matplotlib inline

data=pd.read_csv("../input/corona-virus-report/covid_19_clean_complete.csv")
```

```python
data.rename({'Country/Region':'Country'},axis=1,inplace=True)
```

```python
data['Country'].value_counts().sort_values()
```

```python
data.plot(x='Date',y='Confirmed') #plot between date and confirmed
F = plt.gcf()
Size = F.get_size_inches()
F.set_size_inches(Size[0]*2, Size[1]*1, forward=True)
```

```python
data.plot(x='Date',y='Deaths') #plot between date and no of deaths
plt.xlabel('Date', fontsize=18)
plt.ylabel('Number of deaths', fontsize=16)
F = plt.gcf()
Size = F.get_size_inches()
F.set_size_inches(Size[0]*2.5, Size[1]*1.5, forward=True) 
```

```python
data.plot(x='Date',y='Recovered') #plot between date and recovered number
plt.xlabel('Date', fontsize=18)
plt.ylabel('Recovered', fontsize=16)
F = plt.gcf()
Size = F.get_size_inches()
F.set_size_inches(Size[0]*2, Size[1]*1.5, forward=True) 
```

```python
country_list=data.groupby(['Country']).max().reset_index(drop=None)
print(country_list[['Country','Confirmed','Deaths','Recovered']])
```

```python
country_list=data.groupby(['Lat']).max().reset_index(drop=None)
print(country_list[['Lat','Confirmed','Deaths','Recovered']])
```

```python
country_list=data.groupby(['Long']).max().reset_index(drop=None)
print(country_list[['Long','Confirmed','Deaths','Recovered']])
```

```python
data.plot(x='Country',y='Recovered') #plot between Country and recovered number
plt.xlabel('Country', fontsize=18)
plt.ylabel('Recovered', fontsize=16)
F = plt.gcf()
Size = F.get_size_inches()
F.set_size_inches(Size[0]*3, Size[1]*1.5, forward=True) 
```

```python
plt.plot('Date', 'Deaths', data=data.groupby(['Date']).sum().reset_index(drop=None), color='red')
plt.plot('Date', 'Recovered', data=data.groupby(['Date']).sum().reset_index(drop=None), color='blue')
plt.xticks(rotation=60)
plt.ylabel('Number of cases',fontsize=10)
plt.xlabel('Dates',fontsize=10)
F = plt.gcf()
Size = F.get_size_inches()
F.set_size_inches(Size[0]*3, Size[1]*2, forward=True) 
plt.legend()
plt.show()
```
